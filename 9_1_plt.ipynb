```python
import pandas as pd
import matplotlib.pyplot as plt


def book_plots(file_name):
    data = pd.read_csv(file_name)
    data_2023 = data[data[''] == 2023]
    data_2024 = data[data[''] == 2024]
    plt.figure(figsize=(8, 6))
    line_chart(data_2023)
    bar_chart(data_2024)
    plt.tight_layout()
    plt.show()

def line_chart(data_2023):
    plt.subplot(2,1,1)
    plt.plot(data_2023[''], data_2023[''], 'b^--', label=' ')
    plt.xlabel('')
    plt.ylabel(' ')
    year_2023 = data_2023[''].unique()[0]
    plt.title(f'  {year_2023}')
    plt.legend()
    plt.grid(True)

def bar_chart(data_2024):

    plt.subplot(2, 1, 2)
    plt.bar(data_2024[''], data_2024[''], color='yellow', label=' ')
    plt.xlabel('')
    plt.ylabel(' ')
    year_2024 = data_2024[''].unique()[0]
    plt.title(f'  {year_2024}')
    plt.legend()
    plt.grid(True)

```
