```python
import pandas as pd
import matplotlib.pyplot as plt


def book_plots(file_name):
    data = pd.read_csv(file_name)
    data_2023 = data[data['Год'] == 2023]
    data_2024 = data[data['Год'] == 2024]
    plt.figure(figsize=(8, 6))
    line_chart(data_2023)
    bar_chart(data_2024)
    plt.tight_layout()
    plt.show()

def line_chart(data_2023):
    plt.subplot(2,1,1)
    plt.plot(data_2023['Месяц'], data_2023['Книги'], 'b^--', label='Прочитано книг')
    plt.xlabel('Месяцы')
    plt.ylabel('Количество книг')
    year_2023 = data_2023['Год'].unique()[0]
    plt.title(f'Прочитанные книги {year_2023}')
    plt.legend()
    plt.grid(True)

def bar_chart(data_2024):

    plt.subplot(2, 1, 2)
    plt.bar(data_2024['Месяц'], data_2024['Страницы'], color='yellow', label='Прочитано страниц')
    plt.xlabel('Месяцы')
    plt.ylabel('Количество страниц')
    year_2024 = data_2024['Год'].unique()[0]
    plt.title(f'Прочитанные страницы {year_2024}')
    plt.legend()
    plt.grid(True)

```
