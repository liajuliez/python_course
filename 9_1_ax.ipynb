```python
import pandas as pd
import matplotlib.pyplot as plt

def book_plots(file_name):
    data = pd.read_csv(file_name)
    data_2023 = data[data[''] == 2023]
    data_2024 = data[data[''] == 2024]
    fig, ax = plt.subplots(2, 1, figsize=(8, 6))
    line_chart(data_2023, ax[0])
    bar_chart(data_2024, ax[1])
    plt.tight_layout()
    plt.show()

def line_chart(data_2023, ax):
    ax.plot(data_2023[''], data_2023[''], 'bs-', label=' ')
    ax.set_xlabel('')
    ax.set_ylabel(' ')
    year_2023 = data_2023[''].unique()[0]
    ax.set_title(f'  {year_2023}')
    ax.legend()
    ax.grid(True)

def bar_chart(data_2024, ax):
    ax.bar(data_2024[''], data_2024[''], color='yellow', label=' ')
    ax.set_xlabel('')
    ax.set_ylabel(' ')
    year_2024 = data_2024[''].unique()[0]
    ax.set_title(f'  {year_2024}')
    ax.legend()
    ax.grid(True)

```
