```python
import pandas as pd
import matplotlib.pyplot as plt

def book_plots(file_name):
    data = pd.read_csv(file_name)
    data_2023 = data[data['Год'] == 2023]
    data_2024 = data[data['Год'] == 2024]
    fig, ax = plt.subplots(2, 1, figsize=(8, 6))
    line_chart(data_2023, ax[0])
    bar_chart(data_2024, ax[1])
    plt.tight_layout()
    plt.show()

def line_chart(data_2023, ax):
    ax.plot(data_2023['Месяц'], data_2023['Книги'], 'bs-', label='Прочитано книг')
    ax.set_xlabel('Месяцы')
    ax.set_ylabel('Количество книг')
    year_2023 = data_2023['Год'].unique()[0]
    ax.set_title(f'Прочитанные книги {year_2023}')
    ax.legend()
    ax.grid(True)

def bar_chart(data_2024, ax):
    ax.bar(data_2024['Месяц'], data_2024['Страницы'], color='yellow', label='Прочитано страниц')
    ax.set_xlabel('Месяцы')
    ax.set_ylabel('Количество страниц')
    year_2024 = data_2024['Год'].unique()[0]
    ax.set_title(f'Прочитанные страницы {year_2024}')
    ax.legend()
    ax.grid(True)

```
